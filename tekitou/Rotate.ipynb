```python
import sympy as sym
from sympy import sin as sin
from sympy import cos as cos

psi,theta,phi = sym.symbols("psi theta phi")

R = sym.Matrix([[cos(psi)*cos(theta),cos(psi)*sin(theta)*sin(phi)-cos(phi)*sin(psi),sin(psi)*sin(phi)+cos(psi)*cos(phi)*sin(theta)],
                [cos(theta)*sin(psi),cos(psi)*cos(phi)+sin(psi)*sin(theta)*sin(phi),cos(phi)*sin(psi)*sin(theta)-cos(psi)*sin(phi)],
                [-sin(theta),cos(theta)*sin(phi),cos(theta)*cos(phi)]])

Translate = sym.simplify(R@sym.Matrix([3,3,3]))

Eqaul0 = sym.Eq(Translate[0],0)
Eqaul1 = sym.Eq(Translate[1],0)
Eqaul2 = sym.Eq(Translate[2],27)

Ans = sym.solve((Eqaul0,Eqaul1,Eqaul2),theta)

display(Ans)
```
